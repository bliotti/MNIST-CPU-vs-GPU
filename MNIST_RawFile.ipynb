# Manually Parsing Raw MNIST Images

This notebook shows how to read the `train-images-idx3-ubyte` file manually and display the **first image**.

```python
import struct
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Path to the raw MNIST training images file.
# If you've downloaded MNIST via torchvision, you'll find the raw files in:
#    <your path>/data/MNIST/raw/train-images-idx3-ubyte
# or adjust the path to wherever you've placed the file.

filename = 'data/MNIST/raw/train-images-idx3-ubyte'  # Adjust if it's in a sub-directory

# Each MNIST image is 28x28=784 pixels, stored in unsigned byte (0-255)
IMAGE_SIZE = 28 * 28

# Read the file in binary mode
with open(filename, 'rb') as f:
    # 1) Read the file's header (16 bytes):
    #    - magic number (4 bytes)
    #    - number of images (4 bytes)
    #    - rows (4 bytes)
    #    - columns (4 bytes)
    magic, num_images, rows, cols = struct.unpack('>IIII', f.read(16))

    print(f"Magic number: {magic}")
    print(f"Number of images: {num_images}")
    print(f"Image size: {rows}x{cols}")

    # 2) Read the FIRST image
    #    We'll read 28x28=784 bytes of pixel data.
    image_data_1 = np.fromfile(f, dtype=np.uint8, count=IMAGE_SIZE)
    image_data_1 = image_data_1.reshape(rows, cols)

    # Display the FIRST image
    plt.imshow(image_data_1, cmap='gray')
    plt.title("First image from train-images-idx3-ubyte")
    plt.axis('off')
    plt.show()

    # 3) Read the SECOND image
    #    Just read another 28x28=784 bytes.
    image_data_2 = np.fromfile(f, dtype=np.uint8, count=IMAGE_SIZE)
    image_data_2 = image_data_2.reshape(rows, cols)

    # Display the SECOND image
    plt.imshow(image_data_2, cmap='gray')
    plt.title("Second image from train-images-idx3-ubyte")
    plt.axis('off')
    plt.show()
    
    # 3) Read the SECOND image
    #    Just read another 28x28=784 bytes.
    image_data_3 = np.fromfile(f, dtype=np.uint8, count=IMAGE_SIZE)
    image_data_3 = image_data_3.reshape(rows, cols)

    # Display the SECOND image
    plt.imshow(image_data_3, cmap='gray')
    plt.title("Third image from train-images-idx3-ubyte")
    plt.axis('off')
    plt.show()
```
